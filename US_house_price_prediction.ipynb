Factors influencing the house prices across the United States for the next 10 years.

On a higher level there are 4 factors driving the demand in the realty dector, inturn influencing the house prices.

1)DEMOGRAPHICS      2)ECONOMY        3)INTEREST RATES       4)GOVERNMENT POLICIES  


1)DEMOGRAPHICS-
Demographics are the data that describes the composition of a population, such as age, race, gender, income, migration patterns, and population growth. These statistics are an often overlooked but significant factor that affects how real estate is priced and what types of properties are in demand. Major shifts in the demographics of a nation can have a large impact on real estate trends for several decades.

According to the CIA World Factbook,2018 the median age of American citizens is 38.3 years. The population between 21-44 age group is 103 million, these are the potential buyers of residential houses and also the population for 18-20 age group is 1.1 million, these are the millenials and can be potential buyers 10 years down the line. The retiring or retired population is 8.2 million, who are having their life savings and retirement benefits in hand may also be looking for settling down with buying a house or may be a second house in popular vacation areas.  


2)ECONOMY-
Another key factor that affects the value of real estate, in general and house prices in particular is the overall health of the economy. This is generally measured by economic indicators such as the GDP, employment data, manufacturing activity, the prices of goods, etc. Broadly speaking, when the economy is sluggish, so is real estate.In 2020, because of pandemic, whole world's GDP contracted(US gdp contracted by 3.4%) and unemployment rates(US-8.1%) were highest since sub-prime crisis. Hence, the demand was low so, were the house prices.

3)INTEREST RATES-
Interest rates also have a major impact on the real estate markets. Changes in interest rates can greatly influence a person's ability to purchase a residential property. That is because the lower interest rates go, the lower the cost to obtain a mortgage to buy a home will be, which creates a higher demand for real estate, which again pushes prices up. It's important to note that as interest rates rise, the cost to obtain a mortgage increases, thus lowering demand and prices of real estate. Hence, Federal Interest rates have a great influence on residential house prices.

4)GOVERNMENT POLICIES-
Legislation is also another factor that can have a sizable impact on property demand and prices. Tax credits, deductions, and subsidies are some of the ways the government can temporarily boost demand for real estate for as long as they are in place. Being aware of current government incentives can help you determine changes in supply and demand and identify potentially false trends.
For example, in 2009, the U.S. government introduced a first-time homebuyer's tax credit to homeowners in an attempt to jump-start home sales in a sluggish economy (only those who purchased homes between 2008-2010 were eligible). According to the Government Accountability Office, 2.3 Million people took advantage of the tax incentive.This was quite a sizable increase, although temporary, and without knowing the increase was a result of the tax incentive, you may have ended up concluding that the demand for housing was going up based on other factors.

These are the main factors that will influence the house prices in the USA for the upcoming decade, and are within MECE framework. Over and above these many small and allied factors may affect house prices.

--Unemployment- Related to economic growth is unemployment. When unemployment is rising, fewer people will be able to afford a house. But, even the fear of unemployment may discourage people from entering the property market.

--Consumer confidence. Confidence is important for determining whether people want to take the risk of taking out a mortgage. In particular expectations towards the housing market is important; if people fear house prices could fall, people will defer buying.

--Mortgage availability. In the boom years of 1996-2006, many banks were very keen to lend mortgages. They allowed people to borrow large income multiples. Also, banks required very low deposits (e.g. 100% mortgages). This ease of getting a mortgage meant that demand for housing increased as more people were now able to buy. However, since the credit crunch of 2007, banks and building societies struggled to raise funds for lending on the money markets. Therefore, they have tightened their lending criteria requiring a bigger deposit to buy a house. This has reduced the availability of mortgages and demand fell.

--Geographical factors. Many housing markets are highly geographical. For example, national house prices may be falling, but some areas (e.g. New york, Washington) may still see rising prices. Desirable areas can buck market trends as demand is high, and supply limited. For example, houses near good schools or a good rail link may have a significant premium to other areas.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.regressionplots import influence_plot
import statsmodels.formula.api as smf
import numpy as np
```

```python
us=pd.read_csv("C:\\Users\\Lenovo\\Desktop\\USA\\USA_Housing.csv")
us.head()
```

```python
us=us.drop('Address',axis=1)
```

```python
us.columns=['avg_income','house_age','n_room','n_br','area_pop','price']
```

```python
us['price']=us['price'].apply(np.int64)
```

Number of bedrooms is subset of number of rooms, Hence violates MECE principle. So we will drop n_br feature while building the model. Other feature seem to be Homoscedastic, we will build a multiple linear regression model on the dataset.  

```python
us=us.drop('n_br',axis=1)
```

We will standardize the dataset as the features are represented in different units.

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler().fit(us.iloc[:,:-1].values)
scaled_features=ss.transform(us.iloc[:,:-1].values)

```

```python
us_scaled=pd.DataFrame(scaled_features, index=us.index, columns=us.drop('price',axis=1).columns)
us_scaled['price']=us['price'].values
us_scaled
```

```python
#model building
model= smf.ols('price~avg_income+house_age+n_room+area_pop',data=us_scaled).fit()
model.summary()
```

we have got a good R squared value, model is good for deployment

```python
#we will check for influence of each parameter on house price
ml_avg_inc=smf.ols('price~avg_income',data=us_scaled).fit()
print(ml_avg_inc.pvalues, '/n', ml_avg_inc.rsquared, '/n', ml_avg_inc.rsquared_adj)
```

```python
ml_h_age=smf.ols('price~house_age',data=us_scaled).fit()
print(ml_h_age.pvalues, '/n', ml_h_age.rsquared, '/n', ml_h_age.rsquared_adj)
```

```python
ml_room=smf.ols('price~n_room',data=us_scaled).fit()
print(ml_room.pvalues, '/n', ml_room.rsquared, '/n', ml_room.rsquared_adj)
```

```python
ml_pop=smf.ols('price~area_pop',data=us_scaled).fit()
print(ml_pop.pvalues, '/n', ml_pop.rsquared, '/n', ml_pop.rsquared_adj)
```

```python
#residual analysis
import statsmodels.api as sm
qqplot=sm.qqplot(model.resid,line='q')
plt.title("qq plot of residuals")
plt.show()
```

```python
model_influence=model.get_influence()
(c,_)=model_influence.cooks_distance

plt.stem(np.arange(len(us_scaled)),np.round(c,3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
from statsmodels.graphics.regressionplots import influence_plot
influence_plot(model)
plt.show()
```

# model is found to be performing well,  as leverage and cooks distance are all very low

```python
preds=model.predict(us_scaled)
preds
```
